```python
import pandas as pd
import io
from google.colab import files
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
uploaded = files.upload()
```

```python
pib=pd.read_csv(io.BytesIO(uploaded["PIBE.csv"]),sep=',',decimal='.')
pib.shape
```

```python
columnas = ['Estado', '2018', '2019', '2020']
pibb = pib[columnas]
pibb
```

```python
pibb.set_index('Estado', inplace=True)
```

```python
pibb
```

```python
columnas_a_conservar = ['2020']

# Utiliza la notación de índice para seleccionar las columnas que deseas conservar
pib2020 = pib[columnas_a_conservar]
```

```python
pib2020.head()
```

```python
plt.figure(figsize=(10, 6))
pib2020.plot(kind='bar', color='lightsalmon')
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)
plt.title('PIB por Estado en 2020')
plt.ylabel('PIB')
plt.yticks(fontsize=8)
plt.xticks(rotation=60, ha='right',fontsize=8)

plt.show()
```

```python
columnas_a_conservar = ['2019','2020']

# Utiliza la notación de índice para seleccionar las columnas que deseas conservar
pibb = pib[columnas_a_conservar]
```

```python
pibb.head()
```

```python
bar_width = 0.25
group_space = 0.1 # Ajusta este valor para agregar más espacio entre cada grupo de tres barras
index = np.arange(len(pibb['Estado']))

# Graficar las tasas de mortalidad por entidad
plt.figure(figsize=(18, 6))

# Agregar cuadrícula
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5, color='gray')

# Cambiar los colores de las barras
plt.bar(index, pibb['2018'], label='2018', color='c', width=bar_width)
plt.bar(index + bar_width + group_space, pibb['2019'], label='2019', color='limegreen', width=bar_width)
plt.bar(index + 2 * (bar_width + group_space), pibb['2020'], label='2020', color='lightcoral', width=bar_width)

# Otros ajustes
plt.xlabel('Entidad')
plt.ylabel('PIB (millones de pesos)')
plt.title('PIB 2018-2020')
plt.legend()
plt.xticks(index + (1.5 * (bar_width + group_space)), pibb['Estado'], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(15, 10))
pibb.plot(kind='bar')
plt.title('GDP por Estado (2018-2020)')
plt.xlabel('Año')
plt.ylabel('GDP')
plt.legend(title='Estado', bbox_to_anchor=(1, 1), fontsize=8)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
pibb.T.plot(marker='o')
plt.title('GDP por Estado (2018-2020)')
plt.xlabel('Año')
plt.ylabel('GDP')
plt.legend(title='Estado', bbox_to_anchor=(1, 1),fontsize=8)
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.5)
plt.grid(True)
plt.show()
```

```python
pea=pd.read_csv(io.BytesIO(uploaded["pea.csv"]),sep=',',decimal='.')
pea.shape
```

```python
pea.info()
```

```python
column_position_to_delete = 7
pea.drop(pea.columns[column_position_to_delete], axis=1, inplace=True)

# Muestra el DataFrame después de eliminar la columna
print(pea)
```

```python
pea.head()
```

```python
bar_width = 0.15
group_space = 0.01  # Ajusta este valor para agregar más espacio entre cada grupo de barras
index = np.arange(len(pea['Estado']))

# Graficar las tasas de mortalidad por entidad
plt.figure(figsize=(20, 6))

# Agregar cuadrícula
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5, color='gray')

# Cambiar los colores de las barras
plt.bar(index, pea['15 a 19 años'], label='15 a 19 años', color='c', width=bar_width)
plt.bar(index + bar_width + group_space, pea['20 a 29 años'], label='20 a 29 años', color='limegreen', width=bar_width)
plt.bar(index + 2 * (bar_width + group_space), pea['30 a 39 años'], label='30 a 39 años', color='lightcoral', width=bar_width)
plt.bar(index + 3 * (bar_width + group_space), pea['40 a 49 años'], label='40 a 49 años', color='royalblue', width=bar_width)
plt.bar(index + 4 * (bar_width + group_space), pea['50 a 59 años'], label='50 a 59 años', color='orange', width=bar_width)
plt.bar(index + 5 * (bar_width + group_space), pea['60 años y más'], label='60 años y más', color='magenta', width=bar_width)

# Otros ajustes
plt.xlabel('Entidad')
plt.ylabel('Población')
plt.title('Población por Rango de Edad y Estado')
plt.legend()
plt.xticks(index + (3 * (bar_width + 2 * group_space)), pea['Estado'], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
pea['total'] = pea.sum(axis=1)
pea.head()
```

```python
pea_total = pea[['Estado', 'total']]

# Mostrar el DataFrame resultante
print(pea_total)
```

```python
plt.figure(figsize=(18, 6))

# Agregar cuadrícula
plt.grid(which='major', axis='y', linestyle='--', linewidth=0.5, color='gray')

bars = plt.bar(pea_total['Estado'], pea_total['total'], color='c')
# Cambiar los colores de las barras

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval), ha='center', va='bottom', fontsize=8)

# Otros ajustes
plt.xlabel('Entidad')
plt.ylabel('Población Total')
plt.title('Población ocupada por Estado')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
